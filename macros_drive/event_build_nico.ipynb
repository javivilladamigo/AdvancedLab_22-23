```python
# sort arrays of timestamp and create final timestamp list
timetag0 = np.sort(df0['TIMETAG'])
timetag2 = np.sort(df2['TIMETAG'])
#timetag0 = np.array(df0['TIMETAG'])
#timetag2 = np.array(df2['TIMETAG'])
timetag = []

timetag0 = timetag0[0:1000]
timetag2 = timetag2[0:1000]

# set number of entries to check intersection and tolerance (ps)
sect = int(1000)
tol = 100e6

# intersect the two arrays
for i, j in enumerate(timetag0):

    # before element sect, use the first 2*sect entries
    if i <= sect:
        aux = timetag2[0:2*sect]
    else:
        aux = timetag2[i-sect:i+sect] 
    
    # if one of the element of aux is near, store it
    for k in aux:
        #print(j, k, np.abs(j-k))
        if np.abs(j - k) < tol:
            timetag.append(j)

print(np.size(timetag))
```
